```python
import pandas as pd
import seaborn as sb
import numpy as np
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from imblearn.over_sampling import SMOTENC
from collections import Counter
import sklearn.neighbors.base
sys.modules['sklearn.neighbors.base'] = sklearn.neighbors._base
from missingpy import MissForest
```

Label Encoding, keeping NaN values as NaN

```python
df = pd.read_csv("jobCleanedData.csv", index_col=0)
original = df
mask = df.isnull()
df = df.astype(str).apply(LabelEncoder().fit_transform)
df = df.where(~mask, original)
df.head()
```

Impute missing data with MissForest

```python
imputer = MissForest()
X = df.drop("fraudulent", axis=1)
X = pd.DataFrame(imputer.fit_transform(X))
X.head()
```

Convert float to int and rename the columns

```python
X = X.astype('int')
X.columns = ["title","location","telecommuting","has_company_logo","has_questions","employment_type","required_experience",
                   "required_education","industry","function","description_length","company_profile_length","requirements_length","benefits_length"]        
```

```python
X.head()
```

Split data into train and test

```python
y = df['fraudulent']
X_train,X_test,y_train,y_test= train_test_split(X, y, train_size=0.85, stratify = y, random_state=1)
y_train.shape, y_test.shape
```

SMOTENC to create synthetic samples of minority class

```python
counter = Counter(y_train)
print('Before',counter)
smt = SMOTENC(random_state=1, categorical_features=[True, True, True, True, True, True, True, True, True, True, True, False, False, False, False])
X_train_sm, y_train_sm = smt.fit_resample(X_train, y_train)
counter = Counter(y_train_sm)
print('After',counter)
```

```python
X_train_sm.to_csv("X_train.csv")
y_train_sm.to_csv("y_train.csv")
X_test.to_csv("X_test.csv")
y_test.to_csv("y_test.csv")
```
